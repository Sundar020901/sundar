# DATA VISUALISATION

```python
plt.figure(figsize=(8,6))
sns.histplot(df['Humidity'], bins=20, color='red', kde=True)
plt.xlabel('Humidity')
plt.ylabel('Frequency')
plt.title('Humidity Distrubition')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x = df['Wind Speed_km/h'], color='yellow')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Wind Speed Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='weather_condition', data=df, palette='viridis')
plt.xlabel('weather Condition')
plt.ylabel('Count')
plt.title('weather Condition Counts')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Time based analysis

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
```

```python
df['Year'] = df['Date/Time'].dt.year
df['Month'] = df['Date/Time'].dt.month
df['Day'] = df['Date/Time'].dt.day
df['Hour'] = df['Date/Time'].dt.hour
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='Temp_C', data=df.groupby('Month')['Temp_C'].mean().reset_index())
plt.title('Average Temperature Over Months')
plt.xlabel('Month')
plt.ylabel('Average Temperature (Â°C)')
plt.show()
```

# Correlation Analysis:

```python
print(df.columns)
column_name = 'weather_condition'
if column_name in df.columns:
    df_encoded = pd.get_dummies(df, columns=[column_name])
    correlation_matrix = df_encoded.corr()
    print(correlation_matrix)
else:
    print(f"Column '{column_name}' not found in the DataFrame.")
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5, annot_kws={"size": 8})
plt.title('Correlation Matrix')
plt.show()
```

# Wind Speed Analysis:

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Wind Speed_km/h'], bins=20, color='green', kde=True)
plt.title('Distribution of Wind Speeds')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date/Time', y='Wind Speed_km/h',palette = 'purple', data=df)
plt.title('Wind Speed Over Time')
plt.xlabel('Date/Time')
plt.ylabel('Wind Speed (km/h)')
plt.xticks(rotation=45)
plt.show()
```
